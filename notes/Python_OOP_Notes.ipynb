# Section 1 - Why Did OOP Come Into Existence? 🤔

## The Problem

Imagine you're building a student management system.  
In the **old way** (procedural programming), you'd write something like this:


````python
# ── Old Procedural Way ────────────────────────────────────────────
# Everything is just loose variables and functions floating around

student1_name   = "Arjun"
student1_age    = 20
student1_marks  = 88

student2_name   = "Priya"
student2_age    = 21
student2_marks  = 73

def print_student(name, age, marks):
    print(f"Name: {name}, Age: {age}, Marks: {marks}")

print_student(student1_name, student1_age, student1_marks)
print_student(student2_name, student2_age, student2_marks)
````

### 😣 What's Wrong Here?

- If you have **100 students**, you need 300 variables. That's chaos.
- The data (`name`, `age`, `marks`) and the behaviour (`print_student`) are **not connected** — they're just floating around.
- Anyone can accidentally do `student1_marks = "hello"` and break everything.
- Adding a new feature (like calculating grades) means writing more standalone functions — hard to manage.

## The Real-World Thinking 💡

Think about how we naturally think about the world.

> A **student** *has* a name, age, and marks.  
> A **student** *can* introduce themselves and get their grade.

Everything related to a student is **grouped together**. That's exactly what OOP does.

OOP was born from one simple idea:
> **"Let's model our code the way we model the real world — as objects."**

# Section 2 - Classes and Objects 🏗️

## The Blueprint Analogy

Think of a **class** as an **architect's blueprint** of a house.

- The blueprint defines: number of rooms, door positions, windows, etc.
- From **one blueprint**, you can build **many houses** — each house is different (different colours, furniture) but follows the same design.

```
CLASS  =  Blueprint  (the plan, not the actual thing)
OBJECT =  House      (the real thing built from the plan)
```

## Creating Your First Class


````python
# ── Defining a Class (the blueprint) ─────────────────────────────
class Student:
    
    # __init__ is the "constructor" — it runs automatically
    # when you create a new student object.
    # 'self' means "this particular student"
    
    def __init__(self, name, age, marks):
        self.name  = name    # store name on THIS student
        self.age   = age     # store age  on THIS student
        self.marks = marks   # store marks on THIS student
    
    def introduce(self):
        print(f"Hi! I am {self.name}, {self.age} years old.")
    
    def get_grade(self):
        if self.marks >= 85:
            return "A"
        elif self.marks >= 70:
            return "B"
        else:
            return "C"


# ── Creating Objects (real instances from the blueprint) ──────────
student1 = Student("Arjun", 20, 88)
student2 = Student("Priya", 21, 73)
student3 = Student("Ravi",  19, 55)

# Each object has its OWN data
student1.introduce()
student2.introduce()

print(f"{student1.name}'s grade: {student1.get_grade()}")
print(f"{student2.name}'s grade: {student2.get_grade()}")
print(f"{student3.name}'s grade: {student3.get_grade()}")
````

## Understanding `self`

`self` is just a reference to **the current object**.

When you call `student1.introduce()`, Python internally calls `Student.introduce(student1)`.  
So `self` = `student1` in that moment. It's how each object knows its **own** data.

## Class Variables vs Instance Variables


````python
class Student:
    
    # CLASS variable — shared by ALL students
    school_name = "Python High School"
    total_students = 0
    
    def __init__(self, name, marks):
        # INSTANCE variables — unique to EACH student
        self.name  = name
        self.marks = marks
        Student.total_students += 1   # update the shared count


s1 = Student("Arjun", 88)
s2 = Student("Priya", 73)
s3 = Student("Ravi",  55)

# Instance variable — different for each
print(s1.name)   # Arjun
print(s2.name)   # Priya

# Class variable — same for all
print(s1.school_name)          # Python High School
print(Student.school_name)     # Python High School (access from class too)
print(Student.total_students)  # 3
````

## The 3 Types of Methods

| Method Type | What it does | First parameter |
|---|---|---|
| **Instance method** | Works on a specific object | `self` |
| **Class method** | Works on the class itself | `cls` |
| **Static method** | Just a utility function, no object/class needed | nothing special |


````python
class Student:
    school = "Python High School"
    
    def __init__(self, name, marks):
        self.name  = name
        self.marks = marks
    
    # 1. Instance method — needs a specific student (self)
    def introduce(self):
        print(f"I am {self.name}, my marks are {self.marks}")
    
    # 2. Class method — works on the class, not one student
    @classmethod
    def get_school_name(cls):
        return cls.school
    
    # 3. Static method — just a helper, doesn't need self or cls
    @staticmethod
    def is_passing(marks):
        return marks >= 40


s = Student("Arjun", 88)

s.introduce()                          # instance method
print(Student.get_school_name())       # class method
print(Student.is_passing(35))          # static method → False
print(Student.is_passing(55))          # static method → True
````

# Section 3 - Pillar 1: Encapsulation 🔒

## What is it?

**Encapsulation** = **Wrapping data + methods together** AND **controlling who can access what**.

### Real-Life Analogy 🏧

Think of an **ATM machine**.
- You can *deposit*, *withdraw*, and *check balance*.
- But you **cannot directly open the machine and take money** — there's a controlled interface.
- The internal mechanism (how money is counted, how records are updated) is **hidden** from you.

That's encapsulation — hide the internals, expose only what's needed.

## Access Levels in Python

| Naming | Access Level | Meaning |
|--------|-------------|---------|
| `name` | Public | Anyone can access |
| `_name` | Protected | "Please don't touch from outside" (convention only) |
| `__name` | Private | Truly hidden — Python renames it internally |


````python
class BankAccount:
    
    def __init__(self, owner, balance):
        self.owner     = owner       # PUBLIC   — anyone can see
        self._bank     = "PyBank"    # PROTECTED — internal use preferred
        self.__balance = balance     # PRIVATE   — hidden from outside
    
    # Controlled way to see balance (getter)
    def get_balance(self):
        return self.__balance
    
    # Controlled way to deposit
    def deposit(self, amount):
        if amount <= 0:
            print("❌ Deposit amount must be positive!")
            return
        self.__balance += amount
        print(f"✅ Deposited ₹{amount}. New balance: ₹{self.__balance}")
    
    # Controlled way to withdraw
    def withdraw(self, amount):
        if amount > self.__balance:
            print("❌ Insufficient funds!")
            return
        self.__balance -= amount
        print(f"✅ Withdrew ₹{amount}. New balance: ₹{self.__balance}")


acc = BankAccount("Arjun", 10_000)

# ✅ These work fine
acc.deposit(5000)
acc.withdraw(3000)
print(f"Balance: ₹{acc.get_balance()}")

# ✅ Public attribute — accessible
print(acc.owner)

# ❌ Private attribute — cannot access directly
try:
    print(acc.__balance)
except AttributeError as e:
    print(f"\n❌ Error: {e}")
    print("   You cannot access __balance directly. Use get_balance() instead.")
````

## Using `@property` - A Cleaner Way to Get/Set Values

Instead of writing `get_balance()` and `set_balance()` as regular methods,  
Python gives us a cleaner approach using `@property`.


````python
class Temperature:
    
    def __init__(self, celsius):
        self.__celsius = celsius   # private
    
    # Getter — called when you read  temp.celsius
    @property
    def celsius(self):
        return self.__celsius
    
    # Setter — called when you write temp.celsius = 50
    @celsius.setter
    def celsius(self, value):
        if value < -273.15:   # absolute zero — physically impossible
            print("❌ Temperature below absolute zero is impossible!")
            return
        self.__celsius = value
    
    @property
    def fahrenheit(self):
        return (self.__celsius * 9/5) + 32


t = Temperature(25)
print(f"Celsius:    {t.celsius}°C")
print(f"Fahrenheit: {t.fahrenheit}°F")

t.celsius = 100          # uses the setter
print(f"Updated:    {t.celsius}°C")

t.celsius = -300         # invalid — rejected by setter
print(f"Still:      {t.celsius}°C")  # unchanged
````

### 💡 Why is this great?

- `t.celsius` looks like a simple attribute, but it's actually a **controlled access**.
- You can add **validation logic** without changing how people use your class.
- External code still writes `t.celsius = 100` — clean and simple.

# Section 4 - Pillar 2: Abstraction 🎭

## What is it?

**Abstraction** = **Showing only what matters, hiding how it works.**

### Real-Life Analogy 🚗

When you drive a car:
- You press **accelerate** → car speeds up.
- You press **brake** → car slows down.

You don't need to know about fuel injection, combustion, gear systems.  
The complex machinery is **abstracted away** — you just use the simple interface.

## Abstract Classes in Python

An **abstract class** is like a *contract*:
> "Any class that inherits from me **must** implement these methods."

It ensures all subclasses follow the same structure.


````python
from abc import ABC, abstractmethod

# Abstract class — cannot be used directly
class Animal(ABC):
    
    def __init__(self, name):
        self.name = name
    
    # abstractmethod = subclass MUST implement this
    @abstractmethod
    def speak(self):
        pass
    
    @abstractmethod
    def move(self):
        pass
    
    # Regular method — shared by all animals (no need to override)
    def breathe(self):
        print(f"{self.name} is breathing... 💨")


# Concrete class — implements all abstract methods
class Dog(Animal):
    def speak(self): print(f"{self.name} says: Woof! 🐶")
    def move(self):  print(f"{self.name} runs on 4 legs 🐾")

class Bird(Animal):
    def speak(self): print(f"{self.name} says: Tweet! 🐦")
    def move(self):  print(f"{self.name} flaps its wings and flies ✈️")

class Fish(Animal):
    def speak(self): print(f"{self.name} blows bubbles 🫧")
    def move(self):  print(f"{self.name} swims with fins 🐠")


# Using the concrete classes
dog  = Dog("Bruno")
bird = Bird("Tweety")
fish = Fish("Nemo")

for animal in [dog, bird, fish]:
    animal.speak()
    animal.move()
    animal.breathe()
    print()
````

````python
# What happens if you try to use the abstract class directly?
try:
    a = Animal("Ghost")
except TypeError as e:
    print(f"❌ Error: {e}")
    print("   You cannot create an Animal directly — it's just a blueprint!")

print()

# What happens if a subclass FORGETS to implement a method?
class Snake(Animal):
    def speak(self): print("Ssss 🐍")
    # Oops! forgot to implement move()

try:
    s = Snake("Kaa")
except TypeError as e:
    print(f"❌ Error: {e}")
    print("   Snake must implement ALL abstract methods!")
````

### 💡 Why is Abstraction Useful?

- You define a **common interface** that all subclasses must follow.
- As a user, you don't worry about the *how* — just the *what*.
- Makes large teams work better: one person defines the structure, others fill in the details.

# Section 5 - Pillar 3: Inheritance 👨‍👩‍👧

## What is it?

**Inheritance** = A child class **inherits** all properties and behaviours of a parent class, and can add its own.

### Real-Life Analogy 🧬

You inherit traits from your parents — eye colour, height, certain habits.  
But you also have your own unique traits.

In code: a `Car` *is-a* `Vehicle`. So `Car` can inherit everything `Vehicle` has, and add car-specific things.

## Types of Inheritance


````python
# ── 1. Single Inheritance ─────────────────────────────────────────
#    One parent → One child

class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    
    def start(self):
        print(f"🚀 {self.brand} engine started!")
    
    def info(self):
        print(f"Brand: {self.brand}, Top Speed: {self.speed} km/h")


class Car(Vehicle):   # Car inherits from Vehicle
    def __init__(self, brand, speed, num_doors):
        super().__init__(brand, speed)   # call parent's __init__
        self.num_doors = num_doors       # car's own extra attribute
    
    def honk(self):
        print(f"📯 {self.brand} goes Beep Beep!")


car = Car("Toyota", 180, 4)
car.start()        # inherited from Vehicle ✅
car.info()         # inherited from Vehicle ✅
car.honk()         # Car's own method ✅
print(f"Doors: {car.num_doors}")
````

````python
# ── 2. Multi-level Inheritance ────────────────────────────────────
#    Grandparent → Parent → Child
#    Like: Animal → Mammal → Dog

class Animal:
    def __init__(self, name):
        self.name = name
    def eat(self):
        print(f"{self.name} is eating 🍖")

class Mammal(Animal):
    def breathe(self):
        print(f"{self.name} breathes air 💨")

class Dog(Mammal):
    def bark(self):
        print(f"{self.name} says Woof! 🐶")


dog = Dog("Rex")
dog.eat()      # from Animal (grandparent) ✅
dog.breathe()  # from Mammal (parent) ✅
dog.bark()     # Dog's own method ✅
````

````python
# ── 3. Multiple Inheritance ───────────────────────────────────────
#    A child inherits from MORE THAN ONE parent
#    Think: a FlyingCar is both a Car AND a Plane

class CanFly:
    def fly(self):
        print("✈️  Flying high in the sky!")

class CanDrive:
    def drive(self):
        print("🚗 Driving on the road!")

class FlyingCar(CanFly, CanDrive):   # inherits from BOTH
    def __init__(self, name):
        self.name = name
    
    def show_off(self):
        print(f"{self.name} can do BOTH:")
        self.fly()
        self.drive()


fc = FlyingCar("AeroCar X1")
fc.show_off()
````

````python
# ── 4. Hierarchical Inheritance ───────────────────────────────────
#    One parent → Multiple children
#    Like: Shape is the parent of Circle, Rectangle, Triangle

class Shape:
    def __init__(self, color):
        self.color = color
    
    def describe(self):
        print(f"I am a {self.color} {self.__class__.__name__}")

class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius
    def area(self):
        import math
        return round(math.pi * self.radius ** 2, 2)

class Rectangle(Shape):
    def __init__(self, color, width, height):
        super().__init__(color)
        self.width  = width
        self.height = height
    def area(self):
        return self.width * self.height

class Triangle(Shape):
    def __init__(self, color, base, height):
        super().__init__(color)
        self.base   = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height


shapes = [
    Circle("red", 5),
    Rectangle("blue", 4, 6),
    Triangle("green", 3, 8)
]

for shape in shapes:
    shape.describe()
    print(f"   Area = {shape.area()}")
````

## `super()` - Calling the Parent

`super()` is the way a child class calls the parent class's method.

It's like a child saying: *"Hey Mom/Dad, do your thing first, then I'll add mine."*


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age  = age
        print(f"  [Person] Created: {name}")
    
    def greet(self):
        return f"Hello, I'm {self.name} and I'm {self.age} years old."

class Employee(Person):
    def __init__(self, name, age, company, salary):
        super().__init__(name, age)   # ← calls Person's __init__ first
        self.company = company
        self.salary  = salary
        print(f"  [Employee] Added company info: {company}")
    
    def greet(self):
        base_greeting = super().greet()   # ← calls Person's greet
        return f"{base_greeting} I work at {self.company}."


print("Creating Employee:")
emp = Employee("Sneha", 28, "TechCorp", 60000)
print()
print(emp.greet())
````

# Section 6 - Pillar 4: Polymorphism 🎨

## What is it?

**Polymorphism** = *Many forms* (from Greek: *poly* = many, *morph* = form).

**Same action, different behaviour** depending on the object.

### Real-Life Analogy 📱

Think of the "+" button on a calculator:
- `2 + 3` → adds numbers → `5`
- `"Hello" + " World"` → joins strings → `"Hello World"`

Same symbol (`+`), different behaviour based on what it's applied to. That's polymorphism!

## Method Overriding - Child Rewrites Parent's Method


````python
class Notification:
    def __init__(self, message):
        self.message = message
    
    def send(self):
        print(f"Sending: {self.message}")


class EmailNotification(Notification):
    def __init__(self, message, email):
        super().__init__(message)
        self.email = email
    
    def send(self):   # OVERRIDES parent's send()
        print(f"📧 Email to {self.email}: {self.message}")


class SMSNotification(Notification):
    def __init__(self, message, phone):
        super().__init__(message)
        self.phone = phone
    
    def send(self):   # OVERRIDES parent's send()
        print(f"📱 SMS to {self.phone}: {self.message}")


class WhatsAppNotification(Notification):
    def __init__(self, message, number):
        super().__init__(message)
        self.number = number
    
    def send(self):   # OVERRIDES parent's send()
        print(f"💬 WhatsApp to {self.number}: {self.message}")


# POLYMORPHISM in action:
# Same .send() call, but different behaviour for each object!

notifications = [
    EmailNotification("Your OTP is 1234", "user@email.com"),
    SMSNotification("Your OTP is 1234", "+91-9876543210"),
    WhatsAppNotification("Your OTP is 1234", "+91-9876543210"),
]

print("Sending all notifications:\n")
for notif in notifications:
    notif.send()   # same call — different result!
````

## Duck Typing - "If it walks like a duck and quacks like a duck…"

Python doesn't require formal inheritance for polymorphism.  
If an object **has the method**, Python will call it — no questions asked.

> *"If it quacks like a duck, treat it like a duck."*


````python
# These classes don't inherit from each other — totally independent!
# But they all have an area() method

class Circle:
    def __init__(self, radius):     self.radius = radius
    def area(self): 
        import math
        return round(math.pi * self.radius**2, 2)

class Rectangle:
    def __init__(self, w, h):       self.w, self.h = w, h
    def area(self):                 return self.w * self.h

class Triangle:
    def __init__(self, b, h):       self.b, self.h = b, h
    def area(self):                 return 0.5 * self.b * self.h


# Python doesn't care what TYPE these objects are.
# It just calls .area() on each — if it has area(), great!

shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]

for shape in shapes:
    print(f"{shape.__class__.__name__:12} → Area = {shape.area()}")
````

## Operator Overloading - Teach Python What `+` Means for Your Object


````python
class ShoppingCart:
    def __init__(self, items=None):
        self.items = items or []
    
    def add(self, item, price):
        self.items.append({"item": item, "price": price})
        return self
    
    def total(self):
        return sum(i["price"] for i in self.items)
    
    # Overloading + : combine two carts
    def __add__(self, other):
        combined = ShoppingCart(self.items + other.items)
        return combined
    
    # Overloading len() : number of items
    def __len__(self):
        return len(self.items)
    
    # Overloading str() : nice print
    def __str__(self):
        lines = [f"  • {i['item']}: ₹{i['price']}" for i in self.items]
        return "\n".join(lines) + f"\n  💰 Total: ₹{self.total()}"


cart1 = ShoppingCart()
cart1.add("Notebook", 120).add("Pen", 30)

cart2 = ShoppingCart()
cart2.add("Eraser", 15).add("Ruler", 45)

print("Cart 1:")
print(cart1)

print("\nCart 2:")
print(cart2)

# Using the overloaded + operator
combined = cart1 + cart2
print(f"\nCombined Cart ({len(combined)} items):")
print(combined)
````

# Section 7 - Special (Dunder) Methods ✨

## What are they?

**Dunder** = **D**ouble **under**score → methods like `__init__`, `__str__`, `__len__`.

They're Python's way of saying:
> "Hey, when someone calls `str(obj)` or `len(obj)` or `obj + other`, run *this* method."

You've already seen `__init__`. Let's explore the most useful ones.

### The Most Important Dunders

| Dunder Method | When it's called | Example |
|---|---|---|
| `__init__` | Object created | `Dog("Rex")` |
| `__str__` | `str(obj)` or `print(obj)` | `print(dog)` |
| `__repr__` | `repr(obj)` — for developers | debugging |
| `__len__` | `len(obj)` | `len(cart)` |
| `__add__` | `obj1 + obj2` | `cart1 + cart2` |
| `__eq__` | `obj1 == obj2` | comparing objects |
| `__lt__` | `obj1 < obj2` | sorting |
| `__contains__` | `item in obj` | `"apple" in basket` |
| `__getitem__` | `obj[index]` | `playlist[0]` |
| `__call__` | `obj()` — calling like a function | `tax_calc(1000)` |


````python
# ── __str__ and __repr__ ──────────────────────────────────────────
# __str__  → friendly, for end users  (called by print())
# __repr__ → detailed, for developers (called in console)

class Book:
    def __init__(self, title, author, pages):
        self.title  = title
        self.author = author
        self.pages  = pages
    
    def __str__(self):
        # What users see when they print the book
        return f'"{self.title}" by {self.author}'
    
    def __repr__(self):
        # What developers see for debugging
        return f'Book(title={self.title!r}, author={self.author!r}, pages={self.pages})'


b = Book("The Alchemist", "Paulo Coelho", 208)

print(b)       # calls __str__
print(repr(b)) # calls __repr__
````

````python
# ── __eq__ and __lt__ — comparing objects ────────────────────────

class Student:
    def __init__(self, name, marks):
        self.name  = name
        self.marks = marks
    
    def __str__(self):
        return f"{self.name} ({self.marks})"
    
    # == comparison
    def __eq__(self, other):
        return self.marks == other.marks
    
    # < comparison (enables sorting!)
    def __lt__(self, other):
        return self.marks < other.marks


s1 = Student("Arjun", 88)
s2 = Student("Priya", 73)
s3 = Student("Ravi",  88)

print(f"s1 == s3? {s1 == s3}")   # True  — same marks
print(f"s1 == s2? {s1 == s2}")   # False — different marks
print(f"s2 < s1?  {s2 < s1}")    # True  — 73 < 88

# sorted() works because we defined __lt__
students = [s1, s2, s3, Student("Kavya", 95)]
ranked   = sorted(students, reverse=True)
print("\n🏆 Ranking:")
for i, s in enumerate(ranked, 1):
    print(f"  {i}. {s}")
````

````python
# ── __len__, __getitem__, __contains__ — make objects feel like lists ─

class Playlist:
    def __init__(self, name):
        self.name  = name
        self.songs = []
    
    def add_song(self, song):
        self.songs.append(song)
    
    # len(playlist)
    def __len__(self):
        return len(self.songs)
    
    # playlist[0], playlist[1] ...
    def __getitem__(self, index):
        return self.songs[index]
    
    # "song" in playlist
    def __contains__(self, song):
        return song in self.songs
    
    def __str__(self):
        return f"Playlist '{self.name}' ({len(self)} songs)"


p = Playlist("My Favourites")
p.add_song("Jai Ho")
p.add_song("Tum Hi Ho")
p.add_song("Kesariya")

print(p)
print(f"Number of songs: {len(p)}")
print(f"First song:      {p[0]}")
print(f"'Tum Hi Ho' in playlist? {'Tum Hi Ho' in p}")
print(f"'Shape of You' in playlist? {'Shape of You' in p}")

# Because __getitem__ is defined, we can even loop!
print("\nAll songs:")
for song in p:
    print(f"  🎵 {song}")
````

````python
# ── __call__ — make an object callable like a function ───────────

class TaxCalculator:
    """
    You can use this like a function: tax_calc(amount)
    But it also remembers its tax rate between calls!
    """
    def __init__(self, rate_percent):
        self.rate = rate_percent / 100
    
    def __call__(self, amount):
        tax  = amount * self.rate
        total = amount + tax
        print(f"  Amount: ₹{amount}")
        print(f"  Tax ({self.rate*100}%): ₹{tax:.2f}")
        print(f"  Total: ₹{total:.2f}")
        return total


gst_18 = TaxCalculator(18)
gst_5  = TaxCalculator(5)

print("Restaurant Bill (18% GST):")
gst_18(500)

print("\nGrocery Bill (5% GST):")
gst_5(800)
````

## Context Managers - `__enter__` and `__exit__`

You've seen `with open("file.txt") as f:` - that uses a context manager.  
It automatically opens AND closes the file, even if an error happens.

You can build your own!


````python
class DatabaseConnection:
    """
    Simulates opening and closing a database connection.
    The 'with' statement ensures the connection ALWAYS closes — no leaks!
    """
    
    def __init__(self, db_name):
        self.db_name = db_name
    
    def __enter__(self):
        print(f"🔌 Connecting to database: {self.db_name}...")
        self.connected = True
        return self          # this is what 'as db' gets
    
    def __exit__(self, error_type, error_value, traceback):
        self.connected = False
        print(f"🔒 Connection to {self.db_name} closed.")
        # Return False = don't suppress errors (let them bubble up)
        return False
    
    def query(self, sql):
        if not self.connected:
            raise RuntimeError("Not connected!")
        print(f"  📊 Running: {sql}")
        return ["result1", "result2"]


# The 'with' block handles open AND close automatically
with DatabaseConnection("students_db") as db:
    results = db.query("SELECT * FROM students")
    print(f"  Got {len(results)} rows")

print("After the block — connection is already closed!")
````

# Section 8 - Decorators 🎀

## What is a Decorator?

A **decorator** is a function that **wraps another function** to add extra behaviour - without changing the original function's code.

### Real-Life Analogy 🎁

Think of a gift wrapper.
- You have a gift (your function).
- You wrap it with decorative paper (the decorator).
- The gift is still the same inside - but now it looks different from the outside.

## How Decorators Work - Step by Step


````python
# Step 1: A plain function
def greet():
    print("Hello!")

greet()
````

````python
# Step 2: A function that WRAPS another function
def shout_wrapper(func):
    def inner():
        print("⬇️  About to call the function...")
        func()              # call the original function
        print("⬆️  Done!")
    return inner            # return the wrapped version

# Manually wrapping:
greet = shout_wrapper(greet)
greet()
````

````python
# Step 3: Using @ syntax — same thing, cleaner!

def shout_wrapper(func):
    def inner():
        print("⬇️  About to call the function...")
        func()
        print("⬆️  Done!")
    return inner

@shout_wrapper       # ← this is EXACTLY the same as:  greet = shout_wrapper(greet)
def greet():
    print("Hello!")

greet()
````

## A Practical Decorator - Logging


````python
import functools

def log_calls(func):
    """Logs every time a function is called."""
    @functools.wraps(func)   # preserves the original function's name
    def wrapper(*args, **kwargs):
        print(f"📝 Calling '{func.__name__}' with args={args}, kwargs={kwargs}")
        result = func(*args, **kwargs)
        print(f"✅ '{func.__name__}' returned: {result}")
        return result
    return wrapper


@log_calls
def add(a, b):
    return a + b

@log_calls
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"


add(5, 3)
print()
greet("Arjun")
print()
greet("Priya", greeting="Namaste")
````

## A Practical Decorator - Timing Functions


````python
import time, functools

def timer(func):
    """Measures how long a function takes to run."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start  = time.perf_counter()
        result = func(*args, **kwargs)
        end    = time.perf_counter()
        print(f"⏱️  '{func.__name__}' took {end - start:.4f} seconds")
        return result
    return wrapper


@timer
def slow_task():
    time.sleep(0.5)   # simulate slow work
    return "Task done!"

@timer
def fast_sum(n):
    return sum(range(n))


slow_task()
fast_sum(10_000_000)
````

## Stacking Decorators

You can apply **multiple decorators** to one function - they stack up!  
They apply from **bottom to top** (closest to the function first).


````python
import functools, time

def log_calls(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"📝 Calling {func.__name__}()")
        return func(*args, **kwargs)
    return wrapper

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start  = time.perf_counter()
        result = func(*args, **kwargs)
        print(f"⏱️  Took {time.perf_counter()-start:.4f}s")
        return result
    return wrapper


@log_calls   # applied second (outer)
@timer       # applied first  (inner)
def calculate(n):
    return sum(range(n))

result = calculate(5_000_000)
print(f"Result: {result}")
````

## Built-in Python Decorators You've Already Seen

| Decorator | Where | What it does |
|---|---|---|
| `@property` | Inside class | Makes a method look like an attribute |
| `@classmethod` | Inside class | Method belongs to the class, not instance |
| `@staticmethod` | Inside class | Utility function, no class/instance access |
| `@abstractmethod` | Inside ABC | Forces subclasses to implement this method |


````python
# Quick recap of built-in decorators in one class

from abc import ABC, abstractmethod

class Animal(ABC):
    
    species_count = 0
    
    def __init__(self, name):
        self.name = name
        Animal.species_count += 1
    
    # @abstractmethod — must be overridden
    @abstractmethod
    def speak(self):
        pass
    
    # @property — access like an attribute (no parentheses)
    @property
    def info(self):
        return f"I am {self.name}"
    
    # @classmethod — knows about the class
    @classmethod
    def how_many(cls):
        return f"Total animals created: {cls.species_count}"
    
    # @staticmethod — just a helper
    @staticmethod
    def kingdom():
        return "Animalia"


class Cat(Animal):
    def speak(self): return f"{self.name} says Meow! 🐱"


c1 = Cat("Whiskers")
c2 = Cat("Luna")

print(c1.speak())
print(c1.info)               # property — no ()
print(Animal.how_many())     # classmethod
print(Animal.kingdom())      # staticmethod
````
