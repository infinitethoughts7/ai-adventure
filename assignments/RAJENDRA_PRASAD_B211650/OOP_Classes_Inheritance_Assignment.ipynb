### Q5. Add Behavior — `apply_discount()`

Add a method `apply_discount(percent)` to your `Book` class that:
- Reduces the price by the given percentage
- Prints: `"Discount applied! New price of 'Atomic Habits': ₹400"`

Also add a method `is_long_read()` that:
- Returns `True` if pages > 400, else `False`

```python
b = Book("Atomic Habits", "James Clear", 320, 500)
b.apply_discount(20)
print(b.is_long_read()) # False

b2 = Book("Python Crash Course", "Eric Matthes", 544, 600)
print(b2.is_long_read()) # True
```

````python
# Your code here
class Book:
    def __init__(self, title, author, pages, price):
        self.title=title
        self.author=author
        self.pages=pages
        self.price=price
    def get_summary(self):
        return f"{self.title} by {self.author} - {self.pages} pages -₹{self.price}"
    def apply_discount(self, percent):
        if percent > 0:
            discount=percent/100*self.price
            print(f"Discount applied! New price of '{self.title}': ₹{self.price-discount}")
        else:
            print("Invalid discount percentage")
    def is_long_read(self):
        if self.pages > 400:
            return True
        else:
            return False
    def __str__(self):
        return f"{self.title} by {self.author} — {self.pages} pages — ₹{self.price}"
b = Book("Atomic Habits", "James Clear", 320, 500)
b.apply_discount(20)
print(b.is_long_read())

b2 = Book("Python Crash Course", "Eric Matthes", 544, 600)
print(b2.is_long_read())
````

### Q7. `BankAccount` Class

Create a `BankAccount` class with:
- **Attributes:** `owner`, `balance` (default 0)
- **Methods:**
 - `deposit(amount)` → adds to balance, prints confirmation
 - `withdraw(amount)` → subtracts from balance BUT prints `"Insufficient funds!"` if amount > balance
 - `get_balance()` → returns a formatted string: `"Rahul's balance: ₹5000"`

```python
acc = BankAccount("Rahul", 10000)
acc.deposit(5000) # Deposited ₹5000. New balance: ₹15000
acc.withdraw(3000) # Withdrew ₹3000. New balance: ₹12000
acc.withdraw(20000) # Insufficient funds! Current balance: ₹12000
print(acc.get_balance()) # Rahul's balance: ₹12000
```

````python
# Your code here
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self, amount):
        if amount > 0:
            self.balance+=amount
            print(f" Deposited ₹{amount}. New balance: ₹{self.balance}")
        else:
            print("invalid deposit  amount")
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance-=amount
            print(f"Withdrew ₹{amount}. New balance: ₹{self.balance}")
        else:
            print(f"!insufficient funds! current balance : ₹{self.balance}")
    def get_balance(self):
        return f"{self.owner}'s balance: ₹{self.balance}"
        
acc = BankAccount("Rahul", 10000)
acc.deposit(5000)
acc.withdraw(3000) 
acc.withdraw(20000) 
print(acc.get_balance()) 
````

### Q9. `Animal` → `Dog` → `Cat`

Create a base class `Animal` with:
- **Attributes:** `name`, `sound`
- **Method:** `speak()` → returns `"<name> says <sound>!"`

Then create two child classes:
- `Dog(Animal)` — automatically sets sound to `"Woof"`, adds a method `fetch()` that returns `"<name> fetches the ball!"`
- `Cat(Animal)` — automatically sets sound to `"Meow"`, adds a method `purr()` that returns `"<name> purrs softly..."`

```python
d = Dog("Bruno")
print(d.speak()) # Bruno says Woof!
print(d.fetch()) # Bruno fetches the ball!

c = Cat("Whiskers")
print(c.speak()) # Whiskers says Meow!
print(c.purr()) # Whiskers purrs softly...
```

````python
# Your code here
class Animal:
    def __init__(self, name, sound):
        self.name=name
        self.sound=sound
    def speak(self):
        return f"{self.name} says {self.sound}!"
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name, "Woof")
    def fetch(self):
        return f"{self.name} fetches the ball!"
class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "Meow")
    def purr(self):
        return f"{self.name} purrs softly..."

d = Dog("Bruno")
print(d.speak()) 
print(d.fetch()) 

c = Cat("Whiskers")
print(c.speak())
print(c.purr())
````

### Q10. Employee Management — Full Hierarchy

Build the Employee system from class:

**Base class `Employee`:**
- Attributes: `name`, `emp_id`, `salary`, `department`
- Methods:
 - `get_details()` → `"[E001] Rahul | Engineering | ₹50000"`
 - `give_raise(amount)` → increases salary (only if amount > 0)
 - `calculate_bonus()` → returns 10% of salary

**Child class `Manager(Employee)`:**
- Extra attribute: `team_size`
- Override `calculate_bonus()` → returns 20% of salary
- New method: `approve_leave(emp_name)` → `"Priya approved leave for Rahul"`

**Child class `Intern(Employee)`:**
- Extra attribute: `mentor`
- Override `calculate_bonus()` → returns 5% of salary
- New method: `get_mentor()` → `"Amit's mentor is Priya"`

Then create a list of 5 employees (mix of Employee, Manager, Intern) and loop through them to print details and bonus for each.

**Expected output format:**
```
[E001] Rahul | Engineering | ₹50000 → Bonus: ₹5000.0
[M001] Priya | Engineering | ₹90000 → Bonus: ₹18000.0
[I001] Amit | Engineering | ₹20000 → Bonus: ₹1000.0
...
```

````python
# Your code here
class Employee:
    def __init__(self, name, emp_id, salary, department):
        self.name = name
        self.emp_id=emp_id
        self.salary=salary
        self.department=department
    def get_details(self):
        return f"[{self.emp_id}] {self.name} | {self.department} | ₹{self.salary}"
    def give_raise(self, amount):
        if amount > 0:
            self.salary+=amount
    def calculate_bonus(self):
        return 0.1*self.salary
class Manager(Employee):
    def __init__(self, name, emp_id, salary, department, team_size):
        super().__init__(name, emp_id, salary, department)
        self.team_size=team_size
    def calculate_bonus(self):
        return 0.2*self.salary
    def approve_leave(self, emp_name):
        print(f"{self.name} approved leave for {emp_name}")
class Intern(Employee):
    def __init__(self, name, emp_id, salary, department, mentor):
        super().__init__(name, emp_id, salary, department)
        self.mentor=mentor
    def calculate_bonus(self):
        return 0.05*self.salary
    def get_mentor(self):
        print(f"{self.name} mentor is {self.mentor}")

emp1 = Employee("Rahul", "E001", 50000, "Engineering")
emp2 = Manager("Priya", "M001", 90000, "Engineering", 5)
emp3 = Intern("Amit", "I001", 20000, "Engineering", "Priya")
emp4 = Employee("Sneha", "E002", 40000, "HR")
emp5 = Manager("Kiran", "M002", 80000, "HR", 3)

employees = [emp1, emp2, emp3, emp4, emp5]

for emp in employees:
    print(f"{emp.get_details()} → Bonus: ₹{emp.calculate_bonus()}")
````

### Q13. Library System

Build a mini library:

**`Book` class:**
- Attributes: `title`, `author`, `is_available` (default True)

**`Library` class:**
- Attribute: `books` (an empty list to start)
- Methods:
 - `add_book(book)` → adds a Book object to the library
 - `show_available()` → prints only books where `is_available` is True
 - `lend_book(title)` → finds the book by title, sets `is_available = False`, prints confirmation. If not found or not available, print an error.
 - `return_book(title)` → finds the book, sets `is_available = True`
 - `search(keyword)` → returns all books where keyword is in the title OR author name (case-insensitive)

```python
lib = Library()
lib.add_book(Book("Atomic Habits", "James Clear"))
lib.add_book(Book("Deep Work", "Cal Newport"))
lib.add_book(Book("Python Crash Course", "Eric Matthes"))

lib.show_available()
lib.lend_book("Deep Work")
lib.show_available() # Deep Work should NOT appear
lib.lend_book("Deep Work") # Should say "not available"
lib.return_book("Deep Work")
lib.search("python") # Should find Python Crash Course
```

````python
# Your code here
class Book:
    def __init__(self, title, author):
        self.title=title
        self.author=author
        self.is_available=True
class Library:
    def __init__(self):
        self.books=[]
    def add_book(self, book):
        self.books.append(book)
    def show_available(self):
        for book in self.books:
            if book.is_available:
                print(f"{book.title} by {book.author} is available")
    def lend_book(self, title):
        for book in self.books:
            if book.title==title and book.is_available:
                book.is_available=False
                print(f"Book {book.title} is lend")
                break
        else:
            print(f"Book {title} is not available")
    def return_book(self, title):
        for book in self.books:
            if book.title==title:
                book.is_available=True
                break
    def search(self, keyword):
        keyword=keyword.lower()
        for book in self.books:
            title=book.title.lower()
            author=book.author.lower()
            if keyword in title or keyword in author:
                print(f"keyword found inbook:{book.title}")
lib = Library()
lib.add_book(Book("Atomic Habits", "James Clear"))
lib.add_book(Book("Deep Work", "Cal Newport"))
lib.add_book(Book("Python Crash Course", "Eric Matthes"))

lib.show_available()
lib.lend_book("Deep Work")
lib.show_available() # Deep Work should NOT appear
lib.lend_book("Deep Work") # Should say "not available"
lib.return_book("Deep Work")
lib.search("python")
                
                
````

### Q14. School Report Card System

Build a system with:

**`Student` class:**
- Attributes: `name`, `roll_no`, `marks` (a dictionary like `{"math": 85, "science": 90, "english": 78}`)
- Methods:
 - `average()` → returns the average of all marks
 - `highest_subject()` → returns the subject with the highest mark
 - `pass_or_fail()` → returns `"Pass"` if average >= 40, else `"Fail"`
 - `__str__()` → nice formatted string

**`Topper(Student)` — inherits from Student:**
- Override `__str__()` to add a at the end
- New method: `motivational_quote()` → returns `"<name> says: Hard work pays off!"`

Create 3 students, find who has the highest average (the topper), and print everyone's report card.

**Expected output format:**
```
--- Report Card ---
Name: Rahul | Roll: 101
Marks: math=85, science=90, english=78
Average: 84.33 | Highest: science
Status: Pass

Name: Priya | Roll: 102 
Marks: math=95, science=92, english=88
Average: 91.67 | Highest: math
Status: Pass
Priya says: Hard work pays off!
```

````python
# Your code here
class Student:
    def __init__(self, name, roll_no, marks):
        self.name=name
        self.roll_no=roll_no
        self.marks=marks
    def average(self):
        marks_sum=0
        for key in self.marks:
            marks_sum+=self.marks[key]
        return marks_sum/len(self.marks)
    def highest_subject(self):
        marks=-1
        for key in self.marks:
            if marks < self.marks[key]:
                subject=key
                marks=self.marks[key]
        return f"| Highest: {subject}"
    def pass_or_fail(self):
        if self.average() >= 40:
            return "Pass"
        else:
            return"Fail"
    def __str__(self):
        return f"Name: {self.name} | Roll_no : {self.roll_no} \nMarks: {self.marks}"
class Topper(Student):
     def __init__(self, name, roll_no, marks):
         super().__init__(name, roll_no, marks)
     def motivational_quote(self):
        return f"{self.name} says: Hard work pays off!"

s1 = Student("Rahul", 101, {"math": 85, "science": 90, "english": 78})
s2 = Student("Priya", 102, {"math": 95, "science": 92, "english": 88})
s3 = Student("Amit", 103, {"math": 60, "science": 70, "english": 65})

print(s1)
print(f"Average: {s1.average()}")
print(s1.highest_subject())
print(s1.pass_or_fail())
print("\n")
print(s2)
print(f"Average: {s2.average()}")
print(s2.highest_subject())
print(s2.pass_or_fail())
````
