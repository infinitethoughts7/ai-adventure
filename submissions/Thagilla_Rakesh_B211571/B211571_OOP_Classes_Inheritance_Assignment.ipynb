### Q7. `BankAccount` Class

Create a `BankAccount` class with:
- **Attributes:** `owner`, `balance` (default 0)
- **Methods:**
 - `deposit(amount)` → adds to balance, prints confirmation
 - `withdraw(amount)` → subtracts from balance BUT prints `"Insufficient funds!"` if amount > balance
 - `get_balance()` → returns a formatted string: `"Rahul's balance: ₹5000"`

```python
acc = BankAccount("Rahul", 10000)
acc.deposit(5000) # Deposited ₹5000. New balance: ₹15000
acc.withdraw(3000) # Withdrew ₹3000. New balance: ₹12000
acc.withdraw(20000) # Insufficient funds! Current balance: ₹12000
print(acc.get_balance()) # Rahul's balance: ₹12000
```

````python
# Your code here
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance=self.balance+amount
        print(f'Deposited ₹{amount}. New balance: ₹{self.balance}')
    def withdraw(self,amount):
        if amount>self.balance:
            print(f'Insufficient funds! Current balance: ₹{self.balance}')
        else:
            self.balance=self.balance-amount
            print(f'Withdrew ₹{amount}. New balance: ₹{self.balance}')
    def get_balance(self):
        return f"{self.owner}'s balance: ₹{self.balance}"
acc = BankAccount("Rahul", 10000)
acc.deposit(5000) # Deposited ₹5000. New balance: ₹15000
acc.withdraw(3000) # Withdrew ₹3000. New balance: ₹12000
acc.withdraw(20000) # Insufficient funds! Current balance: ₹12000
print(acc.get_balance()) # Rahul's balance: ₹12000
````

### Q9. `Animal` → `Dog` → `Cat`

Create a base class `Animal` with:
- **Attributes:** `name`, `sound`
- **Method:** `speak()` → returns `"<name> says <sound>!"`

Then create two child classes:
- `Dog(Animal)` — automatically sets sound to `"Woof"`, adds a method `fetch()` that returns `"<name> fetches the ball!"`
- `Cat(Animal)` — automatically sets sound to `"Meow"`, adds a method `purr()` that returns `"<name> purrs softly..."`

```python
d = Dog("Bruno")
print(d.speak()) # Bruno says Woof!
print(d.fetch()) # Bruno fetches the ball!

c = Cat("Whiskers")
print(c.speak()) # Whiskers says Meow!
print(c.purr()) # Whiskers purrs softly...
```

````python
# Your code here
class Animal:
    def __init__(self,name,sound):
        self.name=name
        self.sound=sound
    def speak(self):
        return f'{self.name} says {self.sound}!'
class Dog(Animal):
    def __init__(self,name):
        super().__init__(name,"Woof")
    def fetch(self):
        return f'{self.name} fetches the ball!'
class Cat(Animal):
    def __init__(self,name):
        super().__init__(name,"Meow")
    def purr(self):
        return f'{self.name} purrs softly ...'
d = Dog("Bruno")
print(d.speak()) # Bruno says Woof!
print(d.fetch()) # Bruno fetches the ball!

c = Cat("Whiskers")
print(c.speak()) # Whiskers says Meow!
print(c.purr()) # Whiskers purrs softly...
````

**Child class `Manager(Employee)`:**

````python
# Your code here
class Employee:
    def __init__(self,name,emp_id,salary,department):
        self.name=name
        self.emp_id=emp_id
        self.salary=salary
        self.department = department
    def get_details(self):
        return (f'[{self.emp_id}] {self.name} | {self.department} | ₹{self.salary}')
    def raise_amount(self,amount):
        if amount>0:
            self.salary=self.salary+amount
        else:
            print("Make sure amount should be greater than 0")
    def calculate_bonus(self):
        return  0.1* self.salary
    
class Manager(Employee):
    def __init__(self,name,emp_id,salary,department,team_size):
        super().__init__(name,emp_id,salary,department)
        self.team_size=team_size
    def calculate_bonus(self):
        return  0.2* self.salary
    def approve_leave(self,emp_name):
        print(f'{self.name} approved leave for {emp_name}')
        
class Intern(Employee):
    def __init__(self,name,emp_id,salary,department,mentor):
        super().__init__(name,emp_id,salary,department)
        self.mentor=mentor
    def calculate_bonus(self):
        return  0.05* self.salary
    def get_mentor(self):
        print(f"{self.name}'s mentor is {self.mentor}'")

employees = [
    Manager("Priya", "M001", 80000, "Engineering", 10),
    Employee("Rahul", "E001", 50000, "Engineering"),
    Intern("Amit", "I001", 20000, "Marketing", "Priya"),
    Manager("Sonia", "M002", 90000, "HR", 5),
    Intern("Sneha", "I002", 22000, "Design", "Sonia")
]  

for e in employees:
    print(f'{e.get_details()} -> Bonus : {e.calculate_bonus()}')
````

### Q13. Library System

Build a mini library:

**`Book` class:**
- Attributes: `title`, `author`, `is_available` (default True)

**`Library` class:**
- Attribute: `books` (an empty list to start)
- Methods:
 - `add_book(book)` → adds a Book object to the library
 - `show_available()` → prints only books where `is_available` is True
 - `lend_book(title)` → finds the book by title, sets `is_available = False`, prints confirmation. If not found or not available, print an error.
 - `return_book(title)` → finds the book, sets `is_available = True`
 - `search(keyword)` → returns all books where keyword is in the title OR author name (case-insensitive)

```python
lib = Library()
lib.add_book(Book("Atomic Habits", "James Clear"))
lib.add_book(Book("Deep Work", "Cal Newport"))
lib.add_book(Book("Python Crash Course", "Eric Matthes"))

lib.show_available()
lib.lend_book("Deep Work")
lib.show_available() # Deep Work should NOT appear
lib.lend_book("Deep Work") # Should say "not available"
lib.return_book("Deep Work")
lib.search("python") # Should find Python Crash Course
```

````python
# Your code here
class Book:
    def __init__(self,title,author,is_available=True):
        self.title=title
        self.author=author
        self.is_available=is_available
class Library:
    def __init__(self):
        self.book=[]
    def add_book(self,book):
        self.book.append(book)
    def show_available(self):
        for b in self.book:
            if b.is_available==True:
                print(f'{b.title} written by {b.author}')
    def lend_book(self, title):
        for b in self.book:
            if b.title == title:
                if b.is_available:
                    b.is_available = False
                    print(f"{b.title} has been lent successfully")
                else:
                    print(f"{b.title} is already borrowed")
                return
        print(f"{title} not found in library")
                
    def return_book(self, title):
        for b in self.book:
            if b.title == title:
                if not b.is_available:
                    b.is_available = True
                    print(f"{b.title} has been returned successfully")
                else:
                    print(f"{b.title} was not borrowed")
                return
        print(f"{title} not found in library")
    def search(self, keyword):
        found = False
        for b in self.book:
            if keyword.lower() in b.title.lower() or keyword.lower() in b.author.lower():
                print(f'{b.title} written by {b.author}')
                found = True
        if not found:
            print("No matching books found")

                
lib = Library()
lib.add_book(Book("Atomic Habits", "James Clear"))
lib.add_book(Book("Deep Work", "Cal Newport"))
lib.add_book(Book("Python Crash Course", "Eric Matthes"))

lib.show_available()
lib.lend_book("Deep Work")
lib.show_available() # Deep Work should NOT appear
lib.lend_book("Deep Work") # Should say "not available"
lib.return_book("Deep Work")
lib.search("python") # Should find Python Crash Course
                
            
        
        
    
````

### Q14. School Report Card System

Build a system with:

**`Student` class:**
- Attributes: `name`, `roll_no`, `marks` (a dictionary like `{"math": 85, "science": 90, "english": 78}`)
- Methods:
 - `average()` → returns the average of all marks
 - `highest_subject()` → returns the subject with the highest mark
 - `pass_or_fail()` → returns `"Pass"` if average >= 40, else `"Fail"`
 - `__str__()` → nice formatted string

**`Topper(Student)` — inherits from Student:**
- Override `__str__()` to add a at the end
- New method: `motivational_quote()` → returns `"<name> says: Hard work pays off!"`

Create 3 students, find who has the highest average (the topper), and print everyone's report card.

**Expected output format:**
```
--- Report Card ---
Name: Rahul | Roll: 101
Marks: math=85, science=90, english=78
Average: 84.33 | Highest: science
Status: Pass

Name: Priya | Roll: 102 
Marks: math=95, science=92, english=88
Average: 91.67 | Highest: math
Status: Pass
Priya says: Hard work pays off!
```

````python
# Your code here
class Student:
    def __init__(self,name,roll_no,marks):
        self.name=name
        self.roll_no=roll_no
        self.marks=marks
    def average(self):
        vals = self.marks.values()
        return sum(vals) / len(vals)
    def highest_subject(self):
        return max(self.marks,key=self.marks.get)
    def pass_or_fail(self):
        return "Pass" if self.average()>=40 else "Fail" 
    def __str__(self):
        marks_str = ", ".join([f"{sub}={mark}" for sub, mark in self.marks.items()])
        return (
            f"Name: {self.name} | Roll: {self.roll_no}\n"
            f"Marks: {marks_str}\n"
            f"Average: {self.average():.2f} | Highest: {self.highest_subject()}\n"
            f"Status: {self.pass_or_fail()}"
        )
        
class Topper(Student):
    def __str__(self):
        return super().__str__()   

    def motivational_quote(self):
        return f"{self.name} says: Hard work pays off!"
        
s1 = Student("Rocky", 101, {"math": 85, "science": 90, "english": 78})
s2 = Student("Shiva", 102, {"math": 95, "science": 92, "english": 88})
s3 = Student("Rakesh", 103, {"math": 70, "science": 65, "english": 60})
students = [s1, s2, s3]  
top = max(students, key=lambda s: s.average())
topper = Topper(top.name, top.roll_no, top.marks)
print("--- Report Card ---")
for s in students:
    if s.roll_no == topper.roll_no:
        print(topper)
        print(topper.motivational_quote())
    else:
        print(s)
    print()
````
