### **Assignment-5:Functions**

```python
#Q1
def is_palindrome(text):
    text = text.lower()
    return text == text[::-1]
print(is_palindrome("Racecar"))  # True
print(is_palindrome("Python"))   # False
```

```python
#Q2
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # 120
```

```python
#Q3
def count_vowels(text, vowels="aeiouAEIOU"):
    count = 0
    for ch in text:
        if ch in vowels:
            count += 1
    return count
print(count_vowels("Hello World"))  # 3
```

```python
#Q4
def stats(*args):
    total = 0
    count = 0
    
    min_val = args[0]
    max_val = args[0]
    
    for num in args:
        total += num
        count += 1
        
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num
    
    avg = total / count
    
    return {
        "sum": total,
        "count": count,
        "average": avg,
        "min": min_val,
        "max": max_val
    }
print(stats(10, 20, 30, 40, 50))
```

```python
#Q5
students = {
    "s1": {"name": "Anjali", "marks": [100, 100, 100], "branch": "CSE"},
    "s2": {"name": "Vaishnavi", "marks": [92, 95, 93], "branch": "CIVIL"},
    "s3": {"name": "Venu", "marks": [90, 95, 98], "branch": "CSE"}
}
top_student = ""
highest_avg = 0
for key in students:
    marks = students[key]["marks"]
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks)
    if avg > highest_avg:
        highest_avg = avg
        top_student = students[key]["name"]
print("Top Student:", top_student)
print("Average:", highest_avg)
```

```python
#Q6
def get_stats(numbers):
    total = 0
    for n in numbers:
        total += n
    avg = total / len(numbers)
    min_val = numbers[0]
    max_val = numbers[0]
    for n in numbers:
        if n < min_val:
            min_val = n
        if n > max_val:
            max_val = n
    return min_val, max_val, avg
# Unpacking
mn, mx, avg = get_stats([10, 20, 30, 40, 50])
print("Min:", mn)
print("Max:", mx)
print("Avg:", avg)
```

### **Assignment-7:OOP — Classes & Objects**

```python
#Q1
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return 2 * (self.width + self.height)
    def is_square(self):
        return self.width == self.height 
r = Rectangle(5, 5)
print(r.area())        # 25
print(r.perimeter())   # 20
print(r.is_square())   # True
```

```python
#Q3
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Invalid withdrawal")
    def get_balance(self):
        return self.__balance
acc = BankAccount("Anjali", 1000)
acc.deposit(500)
acc.withdraw(300)
print(acc.get_balance())
```

```python
#Q4
class Student:    
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks   
    def average(self):
        total = 0
        for m in self.marks:
            total += m
        return total / len(self.marks)   
    def __str__(self):
        return self.name
students = [
    Student("Anjali", [98, 90, 90]),
    Student("Anusha", [92, 95, 88]),
    Student("Venu", [90, 95, 80]),
    Student("Abhilash", [88, 84, 90]),
    Student("Vaishnavi", [80, 95, 92])
]

# Sort by average using lambda
students.sort(key=lambda s: s.average(), reverse=True)

for s in students:
    print(s, s.average())
```

```python
#Q6
class Library:   
    def __init__(self):
        self.books = []
    def add_book(self, title, author):
        self.books.append({"title": title, "author": author})
    def remove_book(self, title):
        for book in self.books:
            if book["title"] == title:
                self.books.remove(book)
                break
    def search_by_author(self, author):
        result = []
        for book in self.books:
            if book["author"] == author:
                result.append(book)
        return result
    def show_all(self):
        for book in self.books:
            print(book["title"], "-", book["author"])
lib = Library()
lib.add_book("Python Basics", "Anjali")
lib.add_book("AI Intro", "Deepika")
lib.add_book("ML Guide", "Anusha")
lib.show_all()
print(lib.search_by_author("Anjali"))
```

### **BONUS CHALLENGE**

```python
class Calculator: 
    def __init__(self):
        self.result = 0
    def add(self, value):
        self.result += value
        return self
    def subtract(self, value):
        self.result -= value
        return self
    def multiply(self, value):
        self.result *= value
        return self
    def divide(self, value):
        if value != 0:
            self.result /= value
        else:
            print("Cannot divide by zero")
        return self
    def reset(self):
        self.result = 0
        return self
    def get_result(self):
        return self.result
    def __str__(self):
        return f"Current Result: {self.result}"
```

```python
calc = Calculator()
print(calc.add(5).multiply(3).subtract(2).get_result())  
print(calc.reset().add(10).divide(2).get_result())  
print(calc)  
```
