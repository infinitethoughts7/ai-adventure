Each class can implement the `__add__` method in a way that makes sense for its data type, leading to different behaviors for the same operator.*

### Q9. `Animal` → `Dog` → `Cat`

Create a base class `Animal` with:
- **Attributes:** `name`, `sound`
- **Method:** `speak()` → returns `"<name> says <sound>!"`

Then create two child classes:
- `Dog(Animal)` — automatically sets sound to `"Woof"`, adds a method `fetch()` that returns `"<name> fetches the ball!"`
- `Cat(Animal)` — automatically sets sound to `"Meow"`, adds a method `purr()` that returns `"<name> purrs softly..."`

```python
d = Dog("Bruno")
print(d.speak()) # Bruno says Woof!
print(d.fetch()) # Bruno fetches the ball!

c = Cat("Whiskers")
print(c.speak()) # Whiskers says Meow!
print(c.purr()) # Whiskers purrs softly...
```

````python
# Your code here
class Animal:
    def __init__(self,name,sound):
        self.name = name
        self.sound = sound
    def speak(self):
        return f"The {self.name} says {self.sound}"
class Dog(Animal):
    def __init__(self,name):
        super().__init__(name,"Woof")
    def fetch(self):
        return f"{self.name} is fetching the ball!"
class Cat(Animal):
    def __init__(self,name):
        super().__init__(name,"Meow")
    def purr(self):
        return f"{self.name} is purrs softly!"
d=Dog("Bruno")
print(d.speak())
print(d.fetch())
c=Cat("Whiskers")
print(c.speak())
print(c.purr())
````

### Q10. Employee Management — Full Hierarchy

Build the Employee system from class:

**Base class `Employee`:**
- Attributes: `name`, `emp_id`, `salary`, `department`
- Methods:
 - `get_details()` → `"[E001] Rahul | Engineering | ₹50000"`
 - `give_raise(amount)` → increases salary (only if amount > 0)
 - `calculate_bonus()` → returns 10% of salary

**Child class `Manager(Employee)`:**
- Extra attribute: `team_size`
- Override `calculate_bonus()` → returns 20% of salary
- New method: `approve_leave(emp_name)` → `"Priya approved leave for Rahul"`

**Child class `Intern(Employee)`:**
- Extra attribute: `mentor`
- Override `calculate_bonus()` → returns 5% of salary
- New method: `get_mentor()` → `"Amit's mentor is Priya"`

Then create a list of 5 employees (mix of Employee, Manager, Intern) and loop through them to print details and bonus for each.

**Expected output format:**
```
[E001] Rahul | Engineering | ₹50000 → Bonus: ₹5000.0
[M001] Priya | Engineering | ₹90000 → Bonus: ₹18000.0
[I001] Amit | Engineering | ₹20000 → Bonus: ₹1000.0
...
```

````python

class Employee:
    def __init__(self, name, emp_id, salary, department):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary
        self.department = department

    def get_details(self):
        return f"[{self.emp_id}] {self.name} | {self.department} | ₹{self.salary}"

    def give_raise(self, amount):
        if amount > 0:
            self.salary += amount

    def calculate_bonus(self):
        return 0.10 * self.salary


class Manager(Employee):
    def __init__(self, name, emp_id, salary, department, team_size):
        super().__init__(name, emp_id, salary, department)
        self.team_size = team_size

    def calculate_bonus(self):
        return 0.20 * self.salary

    def approve_leave(self, emp_name):
        return f"{self.name} approved leave for {emp_name}"


class Intern(Employee):
    def __init__(self, name, emp_id, salary, department, mentor):
        super().__init__(name, emp_id, salary, department)
        self.mentor = mentor

    def calculate_bonus(self):
        return 0.05 * self.salary

    def get_mentor(self):
        return f"{self.name}'s mentor is {self.mentor}"


# 🔹 Create employee list (Polymorphic objects)
employees = [
    Employee("Rahul", "E001", 50000, "Engineering"),
    Manager("Priya", "M001", 90000, "Engineering", 5),
    Intern("Amit", "I001", 20000, "Engineering", "Priya"),
    Employee("Sneha", "E002", 60000, "HR"),
    Manager("Arjun", "M002", 100000, "Sales", 8)
]


# 🔹 Loop and display details + bonus
for emp in employees:
    details = emp.get_details()
    bonus = emp.calculate_bonus()
    print(f"{details} → Bonus: ₹{bonus}")
````

### Q13. Library System

Build a mini library:

**`Book` class:**
- Attributes: `title`, `author`, `is_available` (default True)

**`Library` class:**
- Attribute: `books` (an empty list to start)
- Methods:
 - `add_book(book)` → adds a Book object to the library
 - `show_available()` → prints only books where `is_available` is True
 - `lend_book(title)` → finds the book by title, sets `is_available = False`, prints confirmation. If not found or not available, print an error.
 - `return_book(title)` → finds the book, sets `is_available = True`
 - `search(keyword)` → returns all books where keyword is in the title OR author name (case-insensitive)

```python
lib = Library()
lib.add_book(Book("Atomic Habits", "James Clear"))
lib.add_book(Book("Deep Work", "Cal Newport"))
lib.add_book(Book("Python Crash Course", "Eric Matthes"))

lib.show_available()
lib.lend_book("Deep Work")
lib.show_available() # Deep Work should NOT appear
lib.lend_book("Deep Work") # Should say "not available"
lib.return_book("Deep Work")
lib.search("python") # Should find Python Crash Course
```

````python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_available(self):
        for book in self.books:
            if book.is_available:
                print(f"{book.title} by {book.author}")

    def lend_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.is_available:
                    book.is_available = False
                    print(f"{book.title} has been issued.")
                    return
                else:
                    print(f"{book.title} is not available.")
                    return
        print(f"Book '{title}' not found.")

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                book.is_available = True
                print(f"{book.title} has been returned.")
                return
        print(f"Book '{title}' not found.")

    def search(self, keyword):
        results = []
        keyword = keyword.lower()

        for book in self.books:
            if keyword in book.title.lower() or keyword in book.author.lower():
                results.append(book)

        return results
lib = Library()

lib.add_book(Book("Atomic Habits", "James Clear"))
lib.add_book(Book("Deep Work", "Cal Newport"))
lib.add_book(Book("Python Crash Course", "Eric Matthes"))

print("Available books:")
lib.show_available()

print("\nLending Deep Work...")
lib.lend_book("Deep Work")

print("\nAvailable books after lending:")
lib.show_available()

print("\nTrying to lend Deep Work again:")
lib.lend_book("Deep Work")

print("\nReturning Deep Work...")
lib.return_book("Deep Work")

print("\nSearch results for 'python':")
results = lib.search("python")
for b in results:
    print(f"{b.title} by {b.author}")
````

### Q14. School Report Card System

Build a system with:

**`Student` class:**
- Attributes: `name`, `roll_no`, `marks` (a dictionary like `{"math": 85, "science": 90, "english": 78}`)
- Methods:
 - `average()` → returns the average of all marks
 - `highest_subject()` → returns the subject with the highest mark
 - `pass_or_fail()` → returns `"Pass"` if average >= 40, else `"Fail"`
 - `__str__()` → nice formatted string

**`Topper(Student)` — inherits from Student:**
- Override `__str__()` to add a at the end
- New method: `motivational_quote()` → returns `"<name> says: Hard work pays off!"`

Create 3 students, find who has the highest average (the topper), and print everyone's report card.

**Expected output format:**
```
--- Report Card ---
Name: Rahul | Roll: 101
Marks: math=85, science=90, english=78
Average: 84.33 | Highest: science
Status: Pass

Name: Priya | Roll: 102 
Marks: math=95, science=92, english=88
Average: 91.67 | Highest: math
Status: Pass
Priya says: Hard work pays off!
```

````python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks  # dict

    def average(self):
        return sum(self.marks.values()) / len(self.marks)

    def highest_subject(self):
        return max(self.marks, key=self.marks.get)

    def pass_or_fail(self):
        return "Pass" if self.average() >= 40 else "Fail"

    def __str__(self):
        marks_str = ", ".join(f"{sub}={mark}" for sub, mark in self.marks.items())
        avg = round(self.average(), 2)
        highest = self.highest_subject()

        return (
            f"Name: {self.name} | Roll: {self.roll_no}\n"
            f"Marks: {marks_str}\n"
            f"Average: {avg} | Highest: {highest}\n"
            f"Status: {self.pass_or_fail()}"
        )


class Topper(Student):
    def __str__(self):
        base = super().__str__()
        return base  # keep same format; enhancement is via extra method

    def motivational_quote(self):
        return f"{self.name} says: Hard work pays off!"
students = [
    Student("Rahul", 101, {"math": 85, "science": 90, "english": 78}),
    Student("Priya", 102, {"math": 95, "science": 92, "english": 88}),
    Student("Amit", 103, {"math": 60, "science": 70, "english": 65}),
]

# 🔥 Find topper (max by average)
top_student = max(students, key=lambda s: s.average())

# Convert topper into Topper object
topper = Topper(top_student.name, top_student.roll_no, top_student.marks)

print("--- Report Card ---\n")

for s in students:
    # Replace with Topper object if this is the topper
    if s.roll_no == topper.roll_no:
        s = topper

    print(s)
    
    if isinstance(s, Topper):
        print(s.motivational_quote())

    print()

````

### Q18. Git Scenario Questions

---

**a)** You accidentally committed a file with your password in it. The commit is already done but NOT pushed yet. How do you undo it?

*Your answer:*  
- Use `git reset` to undo the last commit  
- Example:  
  - `git reset --soft HEAD~1` → removes commit but keeps changes staged  
  - `git reset --mixed HEAD~1` → removes commit and unstages changes  
- Remove the sensitive file or data  
- Add it to `.gitignore` if needed  
- Commit again safely  

---

**b)** You and your teammate both edited the same file. When you `git pull`, you get a "merge conflict." In simple terms, what happened and what should you do?

*Your answer:*  
- Both changes conflict because Git cannot decide which version to keep  
- Git marks the conflicting sections in the file  
- You should:
  1. Open the file  
  2. Manually resolve conflicts (choose or combine changes)  
  3. Remove conflict markers (`<<<<<<<`, `=======`, `>>>>>>>`)  
  4. Run `git add` and then `git commit`  

---

**c)** What is the difference between `git pull` and `git fetch`?

*Your answer:*  
- `git fetch` → Downloads changes from remote but does NOT merge them  
- `git pull` → Downloads AND merges changes into your current branch

**d)** Why should you write meaningful commit messages instead of `"fixed stuff"` or `"update"`?

*Your answer:*  
- Helps understand what changes were made  
- Makes debugging and tracking history easier  
- Improves team collaboration
